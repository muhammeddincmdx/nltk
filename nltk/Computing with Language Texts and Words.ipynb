```python
def lexical_diversity(text):
    return len(set(text))/len(text)



# Metinleri bir liste iinde topla
texts = [text1, text2, text3,text4,text5,text6,text7,text8,text9]

# 1'den 10'a kadar olan metinler iin lexical_diversity fonksiyonunu altr
for i in range(1, 10):
    text_name = "text" + str(i)
    
    # Eer metin ad varsa, fonksiyonu ar ve sonucu yazdr
    if text_name in globals():
        current_text = globals()[text_name]
        result = lexical_diversity(current_text)
        print(f"{text_name}: {result}")
    else:
        print(f"{text_name} bulunamad.")
```
